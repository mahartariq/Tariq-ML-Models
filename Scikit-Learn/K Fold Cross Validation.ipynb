```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size=0.3)
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
svm = SVC()
svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```

```python
rf = RandomForestClassifier(n_estimators=40)
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```

```python
# Use KFold for our digits example
```

```python
def get_score(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    return model.score(x_test,y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=3)


score_logistic = []
score_svm = []
score_random_forest = []

for train,test in folds.split(digits.data, digits.target):
    x_train,x_test,y_train,y_test = digits.data[train],digits.data[test],digits.target[train],digits.target[test]
    score_logistic.append(get_score(LogisticRegression(),x_train,x_test,y_train,y_test))
    score_svm.append(get_score(SVC(),x_train,x_test,y_train,y_test))
    score_random_forest.append(get_score(RandomForestClassifier(n_estimators=40),x_train,x_test,y_train,y_test))
```

```python
score_logistic
```

```python
score_random_forest
```

# cross_val_score function

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LogisticRegression(),digits.data,digits.target,cv=5)
```

```python
cross_val_score(SVC(),digits.data,digits.target,cv=5)
```

```python
cross_val_score(RandomForestClassifier(),digits.data,digits.target,cv=4)
```

```python
#tuning
```

```python
score1 = cross_val_score(RandomForestClassifier(n_estimators=10),digits.data,digits.target,cv=5)
np.average(score1)
```

```python
score2 = cross_val_score(RandomForestClassifier(n_estimators=20),digits.data,digits.target,cv=5)
np.average(score2)
```

```python
score3 =cross_val_score(RandomForestClassifier(n_estimators=30),digits.data,digits.target,cv=5)
np.average(score3)
```

```python
score4 =  cross_val_score(RandomForestClassifier(n_estimators=55),digits.data,digits.target,cv=5)
np.average(score4)
```
