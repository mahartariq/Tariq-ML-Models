```python
import pandas as pd 
```

```python
df = pd.read_csv('HR_comma_sep.csv')
```

```python
df.head()
```

```python
df.Department.value_counts()
```

```python
depart_dummies = pd.get_dummies(df.Department)
```

```python
depart_dummies.head()
```

```python
salary_dummies = pd.get_dummies(df.salary)
```

```python
salary_dummies.head()
```

```python
df.drop(['salary','Department'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df2 = pd.concat([salary_dummies,depart_dummies,df],axis=1)
```

```python
df2.head()
```

```python
x = df2.drop('left',axis=1)
```

```python
x.head()
```

```python
y = df2['left']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data,test_data,train_target,test_target = train_test_split(x,y,test_size=0.2)
```

```python
from tensorflow.keras import models
```

```python
model = models.Sequential()
```

```python
from tensorflow.keras.layers import *
```

```python
model.add(Dense(128,activation='relu',input_shape=(train_data.shape[1],)))
model.add(Dense(64,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='rmsprop',metrics=['acc'])
```

```python
model.fit(train_data,train_target,epochs=40)
```

```python
model.evaluate(test_data,test_target)
```
