```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetes_datasets = pd.read_csv('diabetes.csv')
diabetes_datasets.head()
```

```python
diabetes_datasets.shape
```

```python
diabetes_datasets.describe()
```

```python
diabetes_datasets['Outcome'].value_counts()
```

```python
diabetes_datasets.groupby('Outcome').mean()
```

```python
x = diabetes_datasets.drop('Outcome',axis=1)
y = diabetes_datasets['Outcome']
```

```python
x.head()
```

```python
y.head()
```

```python
scaler = StandardScaler()
```

```python
transformed_data = scaler.fit_transform(x)
```

```python
x
```

```python
transformed_data
```

```python
x = transformed_data
```

```python
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, stratify=y, test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model = svm.SVC()
```

```python
model.fit(x_train,y_train)
```

```python
train_prediction = model.predict(x_train)
train_acc = accuracy_score(train_prediction,y_train)
```

```python
print(train_acc)
```

```python
test_prediction = model.predict(x_test)
test_acc = accuracy_score(test_prediction,y_test)
```

```python
print(test_acc)
```

```python
input_data = (5,166,72,19,175,25.8,0.587,41)

input_data_as_numpy_array = np.asarray(input_data)

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = model.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
