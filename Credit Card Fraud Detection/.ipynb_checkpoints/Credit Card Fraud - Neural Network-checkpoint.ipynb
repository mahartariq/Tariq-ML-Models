```python
normal = normal.sample(492)
```

```python
len(normal)
```

```python
new_df = pd.concat([normal,fraud], axis=0)
```

```python
new_df.head()
```

```python
new_df.Class.value_counts()
```

```python
x = new_df.drop('Class',axis=1)
y = new_df['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_features,test_features,train_targets,test_targets = train_test_split(x,y,test_size=0.2,stratify=y)
```

```python
print(len(train_features))
print(len(test_features))
```

```python
train_targets.value_counts()
```

```python
test_targets.value_counts()
```

```python
import tensorflow
from tensorflow.keras import models
```

```python
model = models.Sequential()
```

```python
from tensorflow.keras.layers import *
from keras import regularizers
```

```python
model.add(Dense(4096,activation='relu',input_shape=(train_features.shape[-1],)))
model.add(Dense(120,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
opt = tensorflow.keras.optimizers.Adagrad(learning_rate=0.004)
```

```python
model.compile(optimizer=opt,loss='binary_crossentropy',metrics=['acc'])
```

```python
model.fit(train_features,train_targets,epochs=80,verbose=0)
```

```python
model.evaluate(test_features,test_targets)
```

# Update in Model Archi

```python
from tensorflow.keras import models
network = models.Sequential()
from tensorflow.keras.layers import *
```

```python
network.add(Dense(8192,activation='relu',input_shape=(train_features.shape[-1],)))
network.add(Dense(120,activation='relu'))
network.add(Dense(1,activation='sigmoid'))
opt = tensorflow.keras.optimizers.Adamax(learning_rate=0.002)
network.compile(optimizer=opt,loss='binary_crossentropy',metrics=['acc'])
```

```python
network.fit(train_features,train_targets,epochs=100,verbose=0)
network.evaluate(test_features,test_targets)
```
