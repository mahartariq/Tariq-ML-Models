```python
import pandas as pd
```

```python
df = pd.read_csv('sonar data.csv')
```

```python
dic = {'M':0,'R':1}
```

```python
df.replace({'R':dic},inplace=True)
```

```python
df.head()
```

```python
df.R.value_counts()
```

```python
x = df.drop('R',axis=1)
```

```python
x.head()
```

```python
y = df['R']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data,test_data,train_target,test_target = train_test_split(x,y,test_size=0.2,stratify=y)
```

```python
from tensorflow.keras import models
from tensorflow.keras.layers import *
model = models.Sequential()
```

```python
df.shape
```

```python
train_data.shape
```

```python
model.add(Dense(32,activation='relu',input_dim =60)),
model.add(Dense(8,activation='relu')),
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['acc'])
```

```python
model.fit(train_data,train_target,epochs=50,verbose=1)
```

```python
model.evaluate(test_data,test_target)
```
